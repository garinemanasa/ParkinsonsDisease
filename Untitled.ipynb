```python
import os
import json

# Define the folder path containing JSON files
folder_path = r"C:\Users\garin\Documents\GitHub\ParkinsonsDisease\patients"

# Check if folder exists
if not os.path.exists(folder_path):
    print("Error: The specified folder does not exist.")
else:
    # List all JSON files in the folder
    json_files = [file for file in os.listdir(folder_path) if file.endswith('.json')]

    # Initialize a list to store extracted data
    data_list = []

    # Loop through each JSON file
    for file in json_files:
        file_path = os.path.join(folder_path, file)  # Get full file path
        
        # Open and read JSON file
        with open(file_path, "r", encoding="utf-8") as f:
            data = json.load(f)  # Load JSON data
            data_list.append(data)  # Append data to list

    # Print summary
    print(f"Extracted {len(data_list)} JSON files.")
    
    # Optional: Print first extracted file's data (for preview)
    if data_list:
        print(json.dumps(data_list[0], indent=4))  # Pretty print first JSON file's data
```
